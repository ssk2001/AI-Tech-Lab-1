```python

# --- Crawling Web Pages for Search Engines (Simplified Example) ---
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def crawl_web(start_url, max_depth=2):
    visited = set()
    queue = [(start_url, 0)]  # (URL, depth)

    while queue:
        (url, depth) = queue.pop(0)
        if url not in visited and depth <= max_depth:
            visited.add(url)
            try:
                response = requests.get(url)
                soup = BeautifulSoup(response.text, 'html.parser')

                # Process the content of the page (e.g., extract text, index it)
                print(f"Crawled: {url} (Depth: {depth})")

                # Extract links to other pages
                for link in soup.find_all('a'):
                    href = link.get('href')
                    absolute_url = urljoin(url, href)
                    queue.append((absolute_url, depth + 1))

            except Exception as e:
                print(f"Error crawling {url}: {e}")

# Example usage (replace with a valid starting URL)
start_url = "https://www.example.com/"
crawl_web(start_url)



```

```python
# --- Robotics: Path Planning with BFS (Simplified Example) ---
import networkx as nx

def create_grid_map(rows, cols, obstacles):
    G = nx.Graph()
    for i in range(rows):
        for j in range(cols):
            node = (i, j)
            if node not in obstacles:
                G.add_node(node)
                # Add edges to neighbors (up, down, left, right)
                for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                    neighbor = (i + dx, j + dy)
                    if 0 <= neighbor[0] < rows and 0 <= neighbor[1] < cols and neighbor not in obstacles:
                        G.add_edge(node, neighbor)
    return G

def find_path_bfs(graph, start, goal):
    # Use nx.bfs_tree to generate a BFS tree and extract the shortest path
    bfs_tree = nx.bfs_tree(graph, source=start)
    return nx.shortest_path(bfs_tree, source=start, target=goal)  # Find shortest path within the BFS tree

# Example usage
rows = 5
cols = 5
obstacles = [(1, 1), (2, 2), (3, 3)]  # Example obstacles
grid_map = create_grid_map(rows, cols, obstacles)

start_position = (0, 0)
goal_position = (4, 4)

path = find_path_bfs(grid_map, start_position, goal_position)
if path:
    print("Robot Path (BFS):", path)
else:
    print("No path found.")
```
